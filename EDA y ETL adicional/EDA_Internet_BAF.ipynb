# GRAFICOS

Haremos los graficos con los datos de Banda ancha fija y dejaremos de lado el dial up ya que ese sistema no es tan usado y casi que ni modifica los totales de estos dos valores sumados.

```python
plt.figure(figsize=(12, 8))

# Crea un grÃ¡fico de lÃ­neas para todas las provincias
sns.lineplot(data=Internet_BAF_1, x='AÃ±o', y='Banda ancha fija', hue='Provincia', err_style=None)

plt.title("Tendencia de Banda ancha fija (Todas las Provincias)")
plt.xlabel("AÃ±o")
plt.ylabel("Banda ancha fija")

# Ajusta la leyenda para que no se superponga
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Crea un grÃ¡fico de cajas
sns.boxplot(data=Internet_BAF_1, x='AÃ±o', y='Banda ancha fija',hue= 'AÃ±o', palette='coolwarm')

plt.title("DistribuciÃ³n de Banda ancha fija")
plt.xlabel("AÃ±o")
plt.ylabel("Banda ancha fija")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))

# Crea un grÃ¡fico de cajas
sns.boxplot(data=Internet_BAF_1, x='Provincia', y='Banda ancha fija',hue= 'Provincia', palette='coolwarm')

plt.title("DistribuciÃ³n de Banda ancha fija")
plt.xlabel("Provincia")
plt.ylabel("Banda ancha fija")
plt.xticks(rotation=90)
plt.show()
```

Luego de estos tres graficos con los datos cargados incorrectamente se nos hace imposible llegar a hacer una evaluacion de estos datos, ya que no vemos con claridad lo que puede llegar a suceder asi que procederemos a no tener en cuenta esta informacion para las conclusiones finales
