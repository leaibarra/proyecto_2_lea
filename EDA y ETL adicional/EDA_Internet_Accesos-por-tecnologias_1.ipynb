Visualizamos todos los datos y nos encontramos con que hay una fila que solo dice que los datos que tienen un '*' nos explica que: "Los datos provinciales no coinciden a nivel nacional, ya que se rincorporó información que no contien apertuta a nivel geográfico."

```python
Acceso_por_tecno_1 = Acceso_por_tecno_1.dropna()
```

Ya que nos informan que los datos que tienen '*' no coinciden a nivel nacional procedemos a eliminarlos

```python
Acceso_por_tecno_1 = Acceso_por_tecno_1[Acceso_por_tecno_1['Año'] != '2019 *']
```

```python
Acceso_por_tecno_1['Año'] = Acceso_por_tecno_1['Año'].astype(int)
Acceso_por_tecno_1['Trimestre'] = Acceso_por_tecno_1['Trimestre'].astype(int)
```

```python
Acceso_por_tecno_1.info()
```

```python
plt.figure(figsize=(14, 8))

# Crea un gráfico de cajas
sns.boxplot(data=Acceso_por_tecno_1, x='Provincia', y='Total',hue= 'Provincia', palette='coolwarm')

plt.title("Distribución Total por provincia")
plt.xlabel("Provincia")
plt.ylabel("totales")
plt.xticks(rotation=90)
plt.show()
```

Aqui en este grafico se pueden apreciar los errores de los datos de los que venimmos hablando ya que las principales provincias, con mayor cantidad de gente y recursos supuestamente no tiene accesos a internet, lo que seria totalmente contrario
